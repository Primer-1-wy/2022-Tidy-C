## 1.2 单独提取公众号和游记攻略的旅游产品

```python
# 采用Textrank提取关键词组算法
# 这部分待改进
from textrank4zh import TextRank4Keyword  # 导入textrank4zh模块
import numpy as np
def get_keyphrase(s):
    tr4w = TextRank4Keyword(allow_speech_tags=['n','nr','nr1','nr2','nrf','ns','nsf','nt','nz','nz','nl','ng'])
    tr4w.analyze(text=str(s), lower=True, window=5)  # 文本分析，文本小写，窗口为2
    # 最多5个关键词组，有可能一个也没有。词组在原文中出现次数最少为1。
    phase_list = tr4w.get_keyphrases(keywords_num=5, min_occur_num=1)
    if len(phase_list) ==0:
        return np.nan
    else:
        return phase_list[0]
        

# 游记攻略
Travel_tips = pd.concat([Travel_tips1, Travel_tips2], axis=0)  # 游记攻略
def addstr(s):
    return '旅游攻略-'+str(s)
Travel_tips['语料ID'] = Travel_tips['游记ID'].progress_apply(addstr)
Travel_tips['文本'] = Travel_tips['游记标题'] + '\n'+Travel_tips['正文']
Travel_tips['年份'] = pd.to_datetime(Travel_tips['发布时间']).dt.year
Travel_tips['产品名称'] = Travel_tips['文本'].progress_apply(get_keyphrase)


# 微信公众号文章
Wechat_article = pd.concat([Wechat_article1, Wechat_article2], axis=0)  # 微信公众号文章
def addstr(s):
    return '微信公共号文章-'+str(s)
Wechat_article['语料ID'] = Wechat_article['文章ID'].progress_apply(addstr)
Wechat_article['文本'] = Wechat_article['公众号标题'] + '\n'+Wechat_article['正文']
Wechat_article['年份'] = pd.to_datetime(Wechat_article['发布时间']).dt.year
Wechat_article['产品名称'] = Wechat_article['文本'].progress_apply(get_keyphrase)

```

## 2.1 统计情感得分

```python
from cnsenti import Sentiment
senti = Sentiment(pos='E:/Modeling/外挂/外挂/【2022-4-26更新】C题问题二/data/pos.txt',  #正面词典txt文件相对路径
                  neg='E:/Modeling/外挂/外挂/【2022-4-26更新】C题问题二/data/neg.txt',  #负面词典txt文件相对路径
                  merge=True,             #融合cnsenti自带词典和用户导入的自定义词典
                  encoding='utf-8')      #两txt均为utf-8编码
def emotion_score(s):

    r = senti.sentiment_count(str(s))
    if r['pos']>r['neg']:
        score = (r['pos']-r['neg'])/r['words']
    elif r['pos'] < r['neg']:
        score = (r['pos']-r['neg'])/r['words']
    else :
        score = 0
    return score

all_df['情感得分'] = all_df['文本'].progress_apply(emotion_score)
```

## 2.3 计算产品热度得分
将情感得分和频次相加

```python

# 计算综合得分
year_2018['产品热度总分'] = 0.8*year_2018['出现频次']+200*year_2018['情感得分']+0
year_2019['产品热度总分'] = 0.8*year_2019['出现频次']+200*year_2019['情感得分']+1*10
year_2020['产品热度总分'] = 0.8*year_2020['出现频次']+200*year_2020['情感得分']+2*15
year_2021['产品热度总分'] = 0.8*year_2021['出现频次']+200*year_2021['情感得分']+3*20

year_2018['产品热度'] = year_2018['产品热度总分'].div(np.sum(year_2018['产品热度总分']), axis=0)#化成一个小数 加起来为1
year_2019['产品热度'] = year_2019['产品热度总分'].div(np.sum(year_2019['产品热度总分']), axis=0)
year_2020['产品热度'] = year_2020['产品热度总分'].div(np.sum(year_2020['产品热度总分']), axis=0)
year_2021['产品热度'] = year_2021['产品热度总分'].div(np.sum(year_2021['产品热度总分']), axis=0)

year_2018 = year_2018.sort_values(by="产品热度", ascending=False).reset_index(drop=True)
year_2019 = year_2019.sort_values(by="产品热度", ascending=False).reset_index(drop=True)
year_2020 = year_2020.sort_values(by="产品热度", ascending=False).reset_index(drop=True)
year_2021 = year_2021.sort_values(by="产品热度", ascending=False).reset_index(drop=True)

```

## 2.4 判断产品类型

```python
# 分词
import re
import jieba
stopword_list = [k.strip() for k in open(
    'E://Modeling//外挂//外挂//【2022-4-26更新】C题问题二//stop//cn_stopwords.txt', encoding='utf8').readlines() if k.strip() != '']
def clearTxt(line):
    if line != '':
        line = str(line).strip()
        #去除文本中的英文和数字
        line = re.sub("[a-zA-Z0-9]", "", line)
        #只保留中文、大小写字母
        reg = "[^0-9A-Za-z\u4e00-\u9fa5]"
        line = re.sub(reg, '', line)
        #分词
        segList = jieba.cut(line, cut_all=False)
        segSentence = ''
        for word in segList:
            if word != '\t':
                segSentence += word + " "
    # 去停用词
    wordList = segSentence.split(' ')
    sentence = ''
    for word in wordList:
        word = word.strip()
        if word not in stopword_list:
            if word != '\t':
                sentence += word + " "
    return sentence.strip()


product_hot_score['文本'] = product_hot_score['文本'].progress_apply(clearTxt)
product_hot_score
```

```python
# 景区、酒店、网红景点、民宿、特色餐饮、乡村旅游、文创
def get_product_type(s):
    if '景区' in s:
        return '景区'
    elif '酒店' in s:
        return '酒店'
    elif '餐饮' in s:
        return '特色餐饮'
    elif '景点' in s:
        return '景点'
    elif '民宿' in s:
        return '民宿'
    elif '乡村' in s:
        return '乡村旅游'
    elif '文创' in s:
        return '文创'
    else:
        return '景点'

product_hot_score['产品类型判断文本'] = product_hot_score['语料ID'] +' '+product_hot_score['文本']

product_hot_score['产品类型'] = product_hot_score['产品类型判断文本'].progress_apply(get_product_type)
```

```python
# 去除重复的产品
product_hot_score2 = product_hot_score.drop_duplicates(['产品名称'])
product_hot_score2
```
